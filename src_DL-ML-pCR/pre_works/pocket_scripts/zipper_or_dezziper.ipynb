```python
#Zipper

import os
import zipfile
from tqdm import tqdm

def zip_folder(folder_path, output_zip):

    files_to_zip = []
    total_size = 0

    for root, dirs, files in os.walk(folder_path):
        for file in files:
            file_path = os.path.join(root, file)
            files_to_zip.append(file_path)
            total_size += os.path.getsize(file_path)

    with zipfile.ZipFile(output_zip, 'w', zipfile.ZIP_DEFLATED) as zipf:
        with tqdm(total=total_size, unit='B', unit_scale=True, desc="Compression") as pbar:

            for file_path in files_to_zip:
                arcname = os.path.relpath(file_path, folder_path)
                zipf.write(file_path, arcname)

                pbar.update(os.path.getsize(file_path))

#Use
zip_folder("mes_dossiers_nnunet", "MAMAMAIA.zip")
```

```python
#Dezziper

import os
import zipfile
from tqdm import tqdm

def unzip_file(zip_path, extract_to):

    # Ouvrir le zip
    with zipfile.ZipFile(zip_path, 'r') as zipf:

        # Liste des fichiers
        members = zipf.infolist()

        # Taille totale non compressée
        total_size = sum(member.file_size for member in members)

        # Barre de progression
        with tqdm(total=total_size,
                  unit='B',
                  unit_scale=True,
                  desc="Extraction") as pbar:

            for member in members:

                if member.is_dir():
                  continue

                # Extraction du fichier
                zipf.extract(member, extract_to)

                # Mise à jour de la barre
                pbar.update(member.file_size)

# Exemple d'utilisation
unzip_file("mon_dossier.zip", "dossier_extrait")
```
